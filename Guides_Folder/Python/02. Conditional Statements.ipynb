# Conditional Statements

On this Python notebook it will be showed Python conditional Statements

## if - else statement (single condition)
`
if(condition):
    statement1
else
    statement2
`

```python
if( True ):
    print( 'Hello' )
else:
    print( 'Bye' )
```

## if - elif - else statement (multiple condition)
`
if(condition1):
    statement1
elif(confition2)
    statement2
else
    statement3
`

```python
if( False ):
    print( 'Hello' )
elif( True ):
    print( 'What!?' )
else:
    print( 'Bye' )
```
